**[the relationship](https://pubmed.ncbi.nlm.nih.gov/15590696/) between GC content and protein expression in eukaryotes makes this a key factor in [mRNA vaccine design](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC6446947/); a more effiecent vaccine would have higher GC content in the protein coding region.**

```python
a = blast_spike_gene.aln[0].seq
b = blast_spike_gene.aln[1].seq
from Bio.SeqUtils import GC
print(f"vaccine GC: {GC(a)} %")
print(f"spike protein GC: {GC(b)} %")
```

**I assumed that the TC substiutions was just a more efficient form of codon optimization, increasing GC while simultaneously subsituting codons that would [increase gene expression levels](https://www.pnas.org/content/113/41/E6117); codon biases in different organisms can effect gene expression thus I reasoned that the vaccine would substitute the most commonly found codons in the human genome .**

**Armed with this knowledge I decided to try this strategy in an algorithum**


1.If a virus codon already ended on G or C, copy it to memory

2.If not, replace last nucleotide in codon by a G, see if the amino acid still matches, if so, copy it to memory, else try the same with a C

3.take the resulting codon from memory and compare it to a table of most frequently found codons in the human genome, if it maches do nothing, if it doesn't replace it with more commonly found codon for the same amino acid

4.Otherwise copy as is
    

**I eventually took a shortcut and used the library [python_codon_tables](https://pypi.org/project/python-codon-tables/) which takes data from the [codon usage database](https://www.kazusa.or.jp/codon/)**


```python
import python_codon_tables

def use_most_frequent(species):
    '''
    function to search python_codon_tables for a given species
    returns a frequency table corresponding to the most frequent 
    subsitution for a given nucleotide codon
    
    input: str
    return: dict
    
    '''
    substitutions = {}
    codons_table = python_codon_tables.get_codons_table(species)
    for amino_acid in codons_table:
        frequency_table = codons_table[amino_acid]
        codons = sorted(frequency_table)
        max_frequency = 0
        most_frequent_codon = None
        for codon in codons:
            frequency = frequency_table[codon]
            if frequency > max_frequency:
                max_frequency = frequency
                most_frequent_codon = codon
        for codon in codons:
            substitutions[codon] = most_frequent_codon
    return substitutions

bases = ['T', 'C', 'A', 'G']
codons = [a+b+c for a in bases for b in bases for c in bases]
amino_acids = 'FFLLSSSSYY**CC*WLLLLPPPPHHQQRRRRIIIMTTTTNNKKSSRRVVVVAAAADDEEGGGG'
codon_table = dict(zip(codons, amino_acids))
```

**the following code simply takes the codon input of the aligned virus sequence, modifies it according to the algorithum and counts the number of times that codon matches to the vaccine sequence**
